```python
import cv2
import matplotlib.pyplot as plt
import os
import random

def display_validation_batches(folder_path="/content/runs/detect/train23"):
    """
    Function to load and display random validation batch images from a trained YOLOv8 model.

    :param folder_path: Path to the training results folder where batch images are stored.
    """
    # Ensure the folder exists
    if not os.path.exists(folder_path):
        print("Error: Folder does not exist!")
        return

    # Get all batch images from training logs
    batch_images = [f for f in os.listdir(folder_path) if f.startswith("val_batch") and f.endswith(".jpg")]

    if not batch_images:
        print("No validation batch images found. Make sure you ran model.val() first!")
        return

    # Randomly pick up to 6 images for display
    num_images = min(6, len(batch_images))  # Display up to 6 images
    selected_images = random.sample(batch_images, num_images)

    # Plot the selected images in a grid
    fig, axes = plt.subplots(1, num_images, figsize=(20, 10))

    if num_images == 1:
        axes = [axes]  # Ensure axes is iterable for a single image case

    for i, img_name in enumerate(selected_images):
        img_path = os.path.join(folder_path, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert color for Matplotlib

        axes[i].imshow(img)
        axes[i].axis("off")
        axes[i].set_title(f"Batch: {img_name}")

    plt.show()

# Run the function
display_validation_batches("/content/runs/detect/train23")
```

```python
import cv2
import matplotlib.pyplot as plt
import os
import random

def display_ground_truth_vs_predictions(folder_path="/content/runs/detect/train23/"):
    """
    Displays validation batch images comparing ground truth vs predictions.

    :param folder_path: Path to the training results folder where batch images are stored.
    """
    # Ensure the folder exists
    if not os.path.exists(folder_path):
        print("Error: Folder does not exist!")
        return

    # Get all batch images for ground truth and predictions
    gt_images = [f for f in os.listdir(folder_path) if f.startswith("val_batch") and f.endswith("labels.jpg")]
    pred_images = [f.replace("labels", "pred") for f in gt_images]  # Get corresponding predictions

    if not gt_images or not pred_images:
        print("No validation batch images found. Make sure you ran model.val() first!")
        return

    # Randomly pick up to 3 images for display
    num_images = min(3, len(gt_images))  # Display up to 3 images
    selected_gt_images = random.sample(gt_images, num_images)

    # Create a 2-row grid: (Top row = ground truth, Bottom row = predictions)
    fig, axes = plt.subplots(2, num_images, figsize=(20, 10))

    if num_images == 1:
        axes = [axes]  # Ensure axes is iterable for a single image case

    for i, gt_img_name in enumerate(selected_gt_images):
        gt_img_path = os.path.join(folder_path, gt_img_name)
        # Moved the pred_img_name assignment inside the loop
        pred_img_name = gt_img_name.replace("labels", "pred")
        pred_img_path = os.path.join(folder_path, pred_img_name)

        gt_img = cv2.imread(gt_img_path)
        gt_img = cv2.cvtColor(gt_img, cv2.COLOR_BGR2RGB)

        pred_img = cv2.imread(pred_img_path)
        pred_img = cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB)

        # Top row: Ground Truth
        axes[0, i].imshow(gt_img)
        axes[0, i].axis("off")
        axes[0, i].set_title(f"Ground Truth: {gt_img_name}")

        # Bottom row: Predictions
        axes[1, i].imshow(pred_img)
        axes[1, i].axis("off")
        axes[1, i].set_title(f"Predictions: {pred_img_name}")

    plt.show()

# Run the function
display_ground_truth_vs_predictions("/content/runs/detect/train23")
```

```python
import cv2
import torch
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
from IPython.display import clear_output

# Load YOLO model
model = YOLO('/content/drive/MyDrive/yolov8_trained_model_25epoch.pt')

# Open video
# video_path = "/content/drive/MyDrive/Video_Test_2.mp4"
video_path = 0 # Change to 0 for webcam
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define video writer to save output as MP4
output_path = "Video_Test.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # MP4 codec
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Define fixed colors for bounding boxes (Red, Green, Blue)
fixed_colors = [(0, 0, 255), (255, 0, 0), (0, 255, 0)]  # Red, Green, Blue

# Loop through video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if video ends

    # Run YOLOv8 inference on the frame
    results = model(frame, conf=0.2, iou=0.3)

    # Draw bounding boxes
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
            conf = box.conf[0].item()  # Confidence score
            cls = int(box.cls[0].item())  # Class ID
            label = f"{model.names[cls]}: {conf:.2f}"  # Class label

            # Cycle through Red, Green, and Blue
            color = fixed_colors[cls % 3]

            # Draw bounding box
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Write frame to output video
    out.write(frame)

    # Display frame in Colab
    cv2_imshow(frame)
    clear_output(wait=True)  # Clear the output to display the next frame

# Release resources
cap.release()
out.release()

print(f"Processed video saved as {output_path}")
```
